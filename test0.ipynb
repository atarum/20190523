```
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'C')])
plt.subplot(331)
nx.draw(G, node_size=400, node_color='red', with_labels=True, font_weight='bold')

# small graphs
petersen = nx.petersen_graph()
plt.subplot(332)
nx.draw_spring(petersen, node_size=200, node_color='red', with_labels=True, font_weight='bold')
tutte = nx.tutte_graph()
plt.subplot(333)
nx.draw_spring(tutte, node_size=200, node_color='red', with_labels=True, font_weight='bold')
maze = nx.sedgewick_maze_graph()
tet = nx.tetrahedral_graph()

# classic graphs
K_5 = nx.complete_graph(5)
plt.subplot(334)
nx.draw(K_5, node_size=400, node_color='red', with_labels=True, font_weight='bold')
K_3_5 = nx.complete_bipartite_graph(3, 5)
plt.subplot(335)
nx.draw(K_3_5, node_size=400, node_color='red', with_labels=True, font_weight='bold')
barbell = nx.barbell_graph(10, 10)
plt.subplot(336)
nx.draw(barbell, node_size=400, node_color='red', with_labels=True, font_weight='bold')
lollipop = nx.lollipop_graph(10, 20)
plt.subplot(337)
nx.draw(lollipop, node_size=400, node_color='red', with_labels=True, font_weight='bold')

# random graphs
er = nx.erdos_renyi_graph(100, 0.15)
plt.subplot(338)
nx.draw(er, node_size=400, node_color='red', with_labels=True, font_weight='bold')
ws = nx.watts_strogatz_graph(30, 3, 0.1)
ba = nx.barabasi_albert_graph(100, 5)
plt.subplot(339)
nx.draw(ba, node_size=400, node_color='red', with_labels=True, font_weight='bold')
red = nx.random_lobster(100, 0.9, 0.9)
```
